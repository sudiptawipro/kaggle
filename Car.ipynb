```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
```

```python
df = pd.read_csv(URL)
```

```python
df.columns = ['symboling','normalized-losses','make','fuel-type','aspiration',
             'num-of-doors','body-style','drive-wheels','engine-location','wheel-base',
             'length','width','height','curb-weight','engine-type','num-of-cylinders',
             'engine-size','fuel-system','bore','stroke','compression-ratio',
             'horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```python
df.head()
```

```python
set(df['normalized-losses'])
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
#df['normalized-losses'] = df['normalized-losses'].replace({'?':np.nan})
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])
df['normalized-losses'] = df['normalized-losses'].fillna(value=df['normalized-losses'].median())
set(df['normalized-losses'])
```

```python
df['bore'] = df['bore'].replace({'?':np.nan})
df['bore'] = pd.to_numeric(df['bore'])
df['bore'] = df['bore'].fillna(value=df['bore'].median())
```

```python
df['peak-rpm'] = df['peak-rpm'].replace({'?':np.nan})
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'])
df['peak-rpm'] = df['peak-rpm'].fillna(value=df['peak-rpm'].median())
```

```python
df['stroke'] = df['stroke'].replace({'?':np.nan})
df['stroke'] = pd.to_numeric(df['stroke'])
df['stroke'] = df['stroke'].fillna(value=df['stroke'].median())
```

```python
df['horsepower'] = df['horsepower'].replace({'?':np.nan})
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(value=df['horsepower'].median())
```

```python
df['price'] = df['price'].replace({'?':np.nan})
df['price'] = pd.to_numeric(df['price'])
df['price']= df['price'].fillna(value=df['price'].median())
```

```python
df['symboling']= df['symboling'].astype(str)
```

```python
d1=pd.get_dummies(df[['symboling','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels','engine-location','engine-type','num-of-cylinders','fuel-system']],drop_first=True)
```

```python
df=pd.concat([df,d1],axis=1)
```
